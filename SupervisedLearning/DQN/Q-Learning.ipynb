```python
import gymnasium as gym
import numpy as np
from random import randint
```

```python
env= gym.make("MountainCar-v0")
```

[ambiente](https://gymnasium.farama.org/environments/classic_control/mountain_car/)

```python
def discretizar(valor):
    aux = ((valor - env.observation_space.low)/(env.observation_space.high-env.observation_space.low))*20
    return tuple(aux.astype(np.int32))
```

```python
q_table = np.random.uniform(low=-1, high=1, size=[20,20,3])
```

```python
print(discretizar(env.reset()[0]))
env.close()
```

```python
alfa = 0.1
gamma = 0.95
epsilon = 2
episodios = 5000
lista_recompenzas = []
for episodio in range(episodios):
    estado=discretizar(env.reset()[0])
    final= False
    f1 = False
    f2 = False
    recompensa_total = 0
    while not final:
        if randint(0,10)>epsilon:
            accion = np.argmax(q_table[estado])
        else:
            accion =randint(0,2)
            
        nuevo_estado,recompensa,f1,f2,info = env.step(accion)
        q_table[estado][accion] = q_table[estado][accion] + alfa*(recompensa+gamma*np.max(q_table[discretizar(nuevo_estado)])-q_table[estado][accion])
        estado= discretizar(nuevo_estado)
        recompensa_total+=recompensa
        final= f1 or f2
    lista_recompenzas.append(recompensa_total)
    if (episodio+1)%100==0:
        print("Episodio: "+str(episodio)+"| Recompensa promedio: "+str(np.mean(lista_recompenzas)))

env.close()
```

```python
env= gym.make("MountainCar-v0", render_mode="human")
```

```python
estado=discretizar(env.reset()[0])
final= False
f1 = False
f2 = False
recompensa_total = 0
while not final:
    
    accion = np.argmax(q_table[estado])
      
    nuevo_estado,recompensa,f1,f2,info = env.step(accion)
    estado= discretizar(nuevo_estado)
  
    final= f1 or f2
env.close()
```

Implementar q-Learn utilizando el siguiente ambiente: Pendulo
https://gymnasium.farama.org/environments/classic_control/pendulum/

```python
import gymnasium as gym
import numpy as np
from random import randint
```

```python
env= gym.make("Pendulum-v1")
```

```python
def discretizar(valor):
    aux = ((valor - env.observation_space.low)/(env.observation_space.high-env.observation_space.low))*20
    aux = aux.astype(np.int32)
    
    aux = np.clip(aux, 0, 19)
    
    return tuple(aux)
```

```python
print(discretizar(env.reset()[0]))
env.close()
```

```python
q_table = np.random.uniform(low=-1, high=1, size=[20,20,20,5])
```

```python
def continuos(indice):
    if indice == 0:
        val = -1.99
    elif indice == 1:
        val = -1
    elif indice == 2:
        val = 0
    elif indice == 3:
        val = 1
    else:
        val = 2
    
    return np.array([val], dtype=np.float32)
```

```python
alfa = 0.1
gamma = 0.95
epsilon = 2
episodios = 5000
lista_recompenzas = []
for episodio in range(episodios):
    estado=discretizar(env.reset()[0])
    final= False
    f1 = False
    f2 = False
    recompensa_total = 0
    while not final:
        if randint(0,10)>epsilon:
            accion_idx = np.argmax(q_table[estado])
        else:
            accion_idx = randint(0,4)
        
        accion = continuos(accion_idx)

        nuevo_estado,recompensa,f1,f2,info = env.step(accion)
        q_table[estado][accion_idx] = q_table[estado][accion_idx] + alfa*(recompensa+gamma*np.max(q_table[discretizar(nuevo_estado)])-q_table[estado][accion_idx])
        estado= discretizar(nuevo_estado)
        recompensa_total+=recompensa
        final= f1 or f2
    lista_recompenzas.append(recompensa_total)
    if (episodio+1)%100==0:
        print("Episodio: "+str(episodio)+"| Recompensa promedio: "+str(np.mean(lista_recompenzas)))

env.close()
```

```python
env= gym.make("Pendulum-v1", render_mode="human")
```

```python
estado=discretizar(env.reset()[0])
final= False
f1 = False
f2 = False
recompensa_total = 0
while not final:
    
    accion_idx = np.argmax(q_table[estado])
    accion = continuos(accion_idx)
      
    nuevo_estado,recompensa,f1,f2,info = env.step(accion)
    estado= discretizar(nuevo_estado)
  
    final= f1 or f2
env.close()
```
